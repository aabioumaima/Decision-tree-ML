```python
import pandas as pd
```

```python
df = pd.read_csv('/content/titanic.csv')
df
```

```python
final = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
final
```

```python
final['Age'] = final['Age'].fillna(final['Age'].mean())
final
```

```python
#One hot encoding the Sex values by range of 0 and 1

from sklearn.preprocessing import LabelEncoder

sex_n = LabelEncoder()
```

```python
final['l_sex'] = sex_n.fit_transform(final['Sex'])
final
```

```python
X = final['Survived']
X
```

```python
final = final.drop(['Survived', 'Sex'], axis = 'columns')
final.head()
```

```python
#Create the tree decision model

from sklearn import tree

model = tree.DecisionTreeClassifier()
```

```python
model.fit(final, X)
```

```python
model.predict(final)
```

```python
model.score(final, X)
```
